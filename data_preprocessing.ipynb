```python
import pandas as pd
```

```python
data = pd.read_csv('/content/tmpnvxfpdv1.csv')
```

```python
data.head()
```

```python
data.head()
```

```python
data = data[['Lat','Long','HOUR','OCCURRED_ON_DATE','OFFENSE_DESCRIPTION']]
```

```python
data.head()
```

```python
data.dropna(axis=0, how="any", inplace=True)
```

```python
data.info()
```

```python
data['OFFENSE_DESCRIPTION'].unique()
```

```python
# Define risk categories
high_risk = [
    'HARASSMENT/ CRIMINAL HARASSMENT'
]

# Map each category to its risk level
def assign_risk_category(offense):
    if offense in high_risk:
        return 'High'
    else:
        return None  # For offenses not relevant to women's safety

# Apply the function to create the 'Risk_Category' column
data['Risk_Category'] = data['OFFENSE_DESCRIPTION'].apply(assign_risk_category)

# Filter out rows where 'Risk_Category' is None (i.e., not relevant)
df_filtered = data.dropna(subset=['Risk_Category']).reset_index(drop=True)
```
